```python
import matplotlib.pyplot as plt
import pandas

data=pandas.read_csv("2022_04_23_10_11_23_tlm.txt", sep='\t', header=None, names=["TARGET", "PACKET", "HV_PRES0", "HYD_PRES1", "UNUSED", "PACKET_TIMESECONDS"])
data['PACKET_TIMESECONDS'] -= 1650723085.196415
plt.plot(data['PACKET_TIMESECONDS'], data['HV_PRES0'])
plt.xlabel("Seconds", fontsize=10)
plt.ylabel("PSI")
plt.title("Pressure Sensor 0", fontsize=20)
```
